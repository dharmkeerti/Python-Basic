**1.** What are Type I and Type II errors in hypothesis testing, and how do they impact decision-making?
- **Type I Error:** Occurs when the null hypothesis is true but is incorrectly rejected. This is also called a "false positive."

- **Type II Error:** Occurs when the null hypothesis is false but is incorrectly accepted (not rejected). This is also called a "false negative."

- **Impact:** Type I errors may lead to believing an effect exists when it doesn’t, while Type II errors may lead to missing a real effect. The significance level (α) is the probability of a Type I error, and the power of the test (1−β) relates to avoiding Type II errors. Balancing these errors is vital for robust decision-making in experiments.

**2.** What is the P-value in hypothesis testing, and how should it be interpreted
in the context of the null hypothesis?
- **P-value:** The probability of obtaining test results at least as extreme as the observed results, assuming the null hypothesis is true.

- **Interpretation:** A small p-value (< α) means evidence against the null hypothesis, favoring the alternative hypothesis. A large p-value suggests insufficient evidence to reject the null hypothesis. It does not measure the probability that the null hypothesis is true.

**3.** Explain the difference between a Z-test and a T-test, including when to use
each.
- **Z-test:** Used when sample size is large (n > 30) and population standard deviation is known. Assumes normal distribution of data.

- **T-test:** Used when sample size is small (n ≤ 30) or population standard deviation is unknown. More robust for smaller samples with only sample standard deviation.

- **Use Cases:** Use a Z-test for large samples and known population variance; use a T-test for small samples or unknown variance.

**4.** What is a confidence interval, and how does the margin of error influence its width and interpretation?
- **Confidence Interval (CI):** A range of values, derived from the sample data, that is likely to contain the true population parameter.

- **Margin of Error:** Added/subtracted from the sample estimate; larger margin of error gives a wider CI, reflecting less precision. A smaller margin of error gives a narrower CI and more precision.

- **Interpretation:** The width indicates uncertainty; wider CIs mean more uncertainty about the estimate.

**5.** Describe the purpose and assumptions of an ANOVA test. How does it extend hypothesis testing to more than two groups?
- **Purpose:** ANOVA (Analysis of Variance) tests for differences between means of three or more groups.

- **Assumptions:** Independence of observations, normality within groups, and equal variance across groups.

- **Extension:** Unlike t-tests which only compare two groups, ANOVA compares multiple group means simultaneously by analyzing variance between and within groups.

**6.** Write a Python program to perform a one-sample Z-test and interpret the result for a given dataset.

```python
import numpy as np
from scipy.stats import norm

data = [55, 60, 65, 70, 62, 58, 61, 67]
mu = 60  #population mean
sigma = 5  #population std deviation
n = len(data)
sample_mean = np.mean(data)
z = (sample_mean - mu) / (sigma / np.sqrt(n))
p_value = 2 * (1 - norm.cdf(abs(z)))
print(f"Z-score: {z:.2f}, P-value: {p_value:.4f}")
```

**9.** Write a Python function to calculate and visualize the confidence interval for a sample mean.

```python
import numpy as np
import matplotlib.pyplot as plt
from scipy.stats import norm

def plot_ci(data, confidence=0.95):
    sample_mean = np.mean(data)
    sample_std = np.std(data, ddof=1)
    n = len(data)
    z = norm.ppf(1 - (1 - confidence) / 2)
    margin_error = z * (sample_std / np.sqrt(n))
    ci_lower = sample_mean - margin_error
    ci_upper = sample_mean + margin_error

    plt.errorbar(1, sample_mean, yerr=margin_error, fmt='o')
    plt.title(f'95% Confidence Interval: [{ci_lower:.2f}, {ci_upper:.2f}]')
    plt.show()

data = [55, 60, 65, 70, 62, 58, 61, 67]
plot_ci(data)
```

**10.** Perform a Chi-square goodness-of-fit test using Python to compare observed and expected distributions, and explain the outcome.

```python
import numpy as np
from scipy.stats import chisquare

observed = np.array([20, 30, 50])
expected = np.array([25, 25, 50])

chi2, p = chisquare(f_obs=observed, f_exp=expected)
print(f"Chi-square statistic: {chi2:.2f}, P-value: {p:.4f}")
```

- **Explanation:** If the p-value < 0.05, the observed distribution significantly differs from the expected; otherwise, no significant difference.
