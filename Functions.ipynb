1. What is the difference between a function and a method in Python?
- A function is a block of reusable code that is independent and can be called by its name. It is generally defined outside a class. A method is a function that is associated with an object (an instance of a class). It's defined within a class and is implicitly passed the instance on which it was called (usually as the first argument, self).

  - Example:

     i. Function: len("hello")

     ii. Method: "hello".upper() (Here, upper() is called on the string object "hello").

2. Explain the concept of function arguments and parameters in Python.

  - Example:
    
    def multiply(a, b):  # 'a' and 'b' are parameters return a * b

3. What are the different ways to define and call a function in Python?
- Defining a function:
A function is defined using the def keyword, followed by the function name, parentheses (), and a colon : .

- Calling a function:
Functions can be called using various argument passing techniques:
  - Ways to call a function (passing arguments):
    - Positional Arguments: Arguments are passed in the order they are defined in the function signature.
    - Keyword Arguments: Arguments are passed by explicitly naming the parameter they should be assigned to. This allows you to pass them in any order.
    - Default Arguments: Parameters can have default values, making them optional.
    - Arbitrary Positional Arguments (*args): Allows a function to accept an arbitrary number of positional arguments, which are collected into a tuple.
    - Arbitrary Keyword Arguments (**kwargs): Allows a function to accept an arbitrary number of keyword arguments, which are collected into a dictionary.

5. What are iterators in Python and how do they differ from iterables?
- An iterable is any Python object capable of returning its members one at a time (e.g., lists, tuples, strings). An iterable can be looped over.

  An iterator is an object that represents a stream of data. It must implement the __iter__ method (which returns itself) and the __next__ method (which returns the next element in the sequence and raises StopIteration when finished).

  Difference: An iterable is a container you can loop over; an iterator is the worker object that keeps track of the current position during iteration.

  - Example:

    my_list = [1, 2, 3] # Iterable

    my_iterator = iter(my_list) # Iterator
    print(next(my_iterator))    # Output: 1
    print(next(my_iterator))    # Output: 2

6. Explain the concept of generators in Python and how they are defined.
- A generator is a simple and fast way to create iterators. Unlike normal functions that compute and return a whole result set at once, generators yield values one at a time, pausing their execution state and resuming it when the next value is requested.

  They are defined just like normal functions, but instead of using the return statement, they use the yield keyword.

  - Example:
    
    def count_up_to(max):
    i = 1
    
    while i <= max:
    
    yield i  # Pauses and returns 'i'
    
    i += 1

    counter = count_up_to(3)
    print(next(counter)) # Output: 1

7. What are the advantages of using generators over regular functions?
- The main advantages of generators are:
    
    i. Memory Efficiency: They generate items lazily (on demand) and only store the state of the function, not the entire sequence in memory. This is crucial for very large or infinite sequences.
    
    ii. Performance: They allow processing data as it is generated, improving startup time and reducing latency.
    
    iii. Simplicity: Writing generators is often cleaner and simpler than writing custom iterator classes.

- Example: Generating a million numbers with a list requires huge memory, but a generator function only uses a small, fixed amount of memory.

8. What is a lambda function in Python and when is it typically used?
- A lambda function (or anonymous function) is a small, single-expression function defined using the lambda keyword. It can take any number of arguments but can only have one expression.

  Typical Usage: Lambdas are used when a simple function is required for a short duration, usually as an argument to a higher-order function (like map(), filter(), sorted()).

  - Example: Sorting a list of tuples based on the second item.
    
    data = [(1, 'b'), (3, 'a'), (2, 'c')]
    
    #Uses lambda to define a quick function to extract the second item (index 1)

    print(sorted_data) # Output: [(3, 'a'), (1, 'b'), (2, 'c')]

10. What is the difference between `map()`, `reduce()`, and `filter()` functions in Python?
- These three functions are powerful tools for functional programming in Python, often used with iterables.

    - map():
       - Purpose: Applies a function to each item in an iterable and returns an iterator of the results.

       - Transformation: It transforms elements from one form to another, producing a new iterable of the same length as the input (or the shortest input if multiple).

       - Output: An iterable (map object) where each element is the result of the function applied to the corresponding input element.

    - filter():

      - Purpose: Constructs an iterator from elements of an iterable for which a function returns true.

      - Selection: It filters elements based on a condition, effectively selecting a subset of the original iterable.

      - Output: An iterable (filter object) containing only the elements for which the provided function returned True. The output iterable's length can be less than or equal to the input iterable's length.

    - reduce(): (Note: reduce is in the functools module, not a built-in function)

      - Purpose: Applies a function of two arguments cumulatively to the items of an iterable, from left to right, so as to reduce the iterable to a single value.

      - Aggregation/Reduction: It aggregates or reduces a sequence of values down to a single value.

      - Output: A single, accumulated value.

- Example:
  - map: list(map(lambda x: x * 2, [1, 2, 3])) -> [2, 4, 6]

  - filter: list(filter(lambda x: x > 2, [1, 2, 3, 4])) -> [3, 4]

  - reduce: from functools import reduce / reduce(lambda x, y: x + y, [1, 2, 3, 4]) -> 10

1. Write a Python function that takes a list of numbers as input and returns the sum of all even numbers in the list.

```python
def sum_of_evens(numbers):
    """Takes a list of numbers and returns the sum of all even numbers."""
    total = 0
    for num in numbers:
        if num % 2 == 0:
            total += num
    return total

# Example usage:
my_list = [1, 2, 3, 4, 5, 6, 7, 8]
print(sum_of_evens(my_list))
# Output: 20
```

4. Write a Python function that checks if a given number is prime or not from 1 to 200.

```python
def is_prime(n):
    """Checks if a given number (n) is prime."""
    if n <= 1:
        return False
    # Only need to check divisors up to the square root of n
    for i in range(2, int(n**0.5) + 1):
        if n % i == 0:
            return False
    return True

# Example usage:
number = 17
print(f"{number} is prime: {is_prime(number)}")
# Output: 17 is prime: True
```

5. Create an iterator class in Python that generates the Fibonacci sequence up to a specified number of terms.

```python
class FibonacciIterator:
    """Iterator that generates the Fibonacci sequence up to 'max_terms'."""
    def __init__(self, max_terms):
        self.max_terms = max_terms
        self.count = 0
        self.a = 0
        self.b = 1

    def __iter__(self):
        return self

    def __next__(self):
        if self.count >= self.max_terms:
            raise StopIteration

        if self.count == 0:
            self.count += 1
            return 0
        elif self.count == 1:
            self.count += 1
            return 1
        else:
            self.a, self.b = self.b, self.a + self.b
            self.count += 1
            return self.b

# Example usage:
fib_seq = FibonacciIterator(7)
print(list(fib_seq))
# Output: [0, 1, 1, 2, 3, 5, 8]
```

6. Write a generator function in Python that yields the powers of 2 up to a given exponent.

```python
def power_of_two_generator(exponent):
    """Yields powers of 2 (2**i) up to the specified exponent."""
    for i in range(exponent + 1):
        yield 2 ** i

# Example usage:
powers = power_of_two_generator(5)
print(list(powers))
# Output: [1, 2, 4, 8, 16, 32]
```

7. Implement a generator function that reads a file line by line and yields each line as a string.

```python
def file_line_reader(filepath):
    """Generator that reads a file line by line and yields each line."""
    try:
        with open(filepath, 'r') as file:
            for line in file:
                yield line.rstrip('\n') # Remove newline characters
    except FileNotFoundError:
        print(f"Error: File not found at {filepath}")

# To test, create a temporary file named 'test_file.txt'
# with content: "Line 1\nLine 2\nLine 3"
#
# Example usage:
# for line in file_line_reader('test_file.txt'):
#     print(line)
```

8. Use a lambda function in Python to sort a list of tuples based on the second element of each tuple.

```python
data = [('apple', 3), ('banana', 1), ('cherry', 2)]

# Use lambda function as the key for sorting (x[1] is the second element)
sorted_data = sorted(data, key=lambda x: x[1])

print(sorted_data)
# Output: [('banana', 1), ('cherry', 2), ('apple', 3)]
```

9. Write a Python program that uses `map()` to convert a list of temperatures from Celsius to Fahrenheit.

```python
def celsius_to_fahrenheit(c):
    """Converts Celsius to Fahrenheit: F = (C * 9/5) + 32"""
    return (c * 9/5) + 32

celsius_temps = [0, 10, 20, 30, 40]

# Use map() to apply the conversion function to every element
fahrenheit_temps = list(map(celsius_to_fahrenheit, celsius_temps))

print(fahrenheit_temps)
# Output: [32.0, 50.0, 68.0, 86.0, 104.0]
```

10. Create a Python program that uses `filter()` to remove all the vowels from a given string.

```python
def is_not_vowel(char):
    """Returns True if the character is not a vowel."""
    vowels = 'aeiouAEIOU'
    return char not in vowels

my_string = "Programming"

# Use filter() with the selection function
consonants_iterator = filter(is_not_vowel, my_string)

# Join the remaining characters back into a string
result_string = "".join(consonants_iterator)

print(result_string)
# Output: Prgrmmng
```
