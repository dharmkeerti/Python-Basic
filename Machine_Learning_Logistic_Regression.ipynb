**Question 1. What is Simple Linear Regression (SLR)? Explain its purpose.**
- Simple Linear Regression is a method used to study the relationship between one independent variable (X) and one dependent variable (Y).
- Its main purpose is to understand how Y changes when X changes, and to predict Y using X.

**Question 2. What are the key assumptions of Simple Linear Regression?**
1. **Linearity** - The relationship between X and Y is linear.
2. **Independence** - The observations are independent of each other.
3. **homoscedasticity** - The variance of errors is constant.
4. **Normality of residuals** -  Errors follow a normal distribution.
5. **No Outliers** - There are no major outliers.


**Question 3.  Write the mathematical equation for a simple linear regression model and
explain each term.**
- Y = β0 + β1X + ε

  - Y = dependent variable (output)
  - X = independent variable (input)
  - β0 = intercept (value of Y when X = 0)
  - β1 = slope (how much Y changes when X increases by 1)
  - ε = error term (unexplained variation)

**Question 4.  Provide a real-world example where simple linear regression can be applied.**
- Example: Predicting house price from house area.

  - X = area (square feet)
  - Y = price

- As area increases, price usually increases, so simple linear regression can predict price for new houses.

**Question 5. What is the method of least squares in linear regression?**
- The method of least squares finds the “best-fit” line by minimizing the sum of squared differences between actual values and predicted values.

- It chooses the line where the total squared error is the smallest.

**Question 6.  What is Logistic Regression? How does it differ from Linear Regression?**
- Logistic Regression is used when the output (dependent variable) is categorical, usually 0 or 1 (for example: pass/fail, yes/no, churn/not churn).
- It predicts the probability that something belongs to a class.

- Difference:

  - Linear Regression predicts continuous values.
  - Logistic Regression predicts probability (between 0 and 1) and is used for classification.

**Question 7. Name and briefly describe three common evaluation metrics for regression models.**
1. MAE (Mean Absolute Error): Average absolute difference between real and predicted values.
2. MSE (Mean Squared Error): Average of squared errors (penalizes large mistakes more).
3. RMSE (Root Mean Squared Error): Square root of MSE, easier to interpret because it is in the same units as the target variable.

**Question 8. What is the purpose of the R-squared metric in regression analysis?**
- R-squared measures how much of the variation in the dependent variable (Y) is explained by the model.

- R-squared ranges from 0 to 1:

  - 0 means the model explains nothing.
  - 1 means the model explains everything.

- Higher R-squared generally means a better fit, but it does not always mean the model is perfect.

**Question 9. Write Python code to fit a simple linear regression model using scikit-learn
and print the slope and intercept.**


```python
# Simple Linear Regression using scikit-learn

from sklearn.linear_model import LinearRegression
import numpy as np

# Sample data
X = np.array([1, 2, 3, 4, 5]).reshape(-1, 1)
Y = np.array([2, 4, 5, 7, 10])

# Create model
model = LinearRegression()

# Fit model
model.fit(X, Y)

# Print slope and intercept
print("Slope:", model.coef_[0])
print("Intercept:", model.intercept_)
```

**Question 10. How do you interpret the coefficients in a simple linear regression model?**
- The slope tells us how much the dependent variable (Y) changes when the independent variable (X) increases by 1 unit.

- If slope = 2, it means:
  - for every 1 unit increase in X, Y increases by 2.

- The intercept is the value of Y when X = 0.

- If intercept = 5, it means:
  - when X is zero, the predicted value of Y is 5.
